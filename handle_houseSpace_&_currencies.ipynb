```python

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/df2020.csv', error_bad_lines=False)
```

```python
#handled different currencies
y = pd.get_dummies(df.priceType, prefix='PriceType')
print(y.head(30))
```

```python
#handle houseSpace MISSING VALUE
houseSpaceNew = df.houseSpace
df.houseSpace = houseSpaceNew.fillna(houseSpaceNew.median())
df.houseSpace.isnull().sum()
```

```python
houseSpaceNew = df.houseSpace
houseSpaceNew 
```

```python
result = pd.concat([houseSpaceNew, y,df.price], axis=1)
```
